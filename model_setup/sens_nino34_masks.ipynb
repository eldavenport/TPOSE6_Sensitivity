```python
import xarray as xr 
import numpy as np 

project_dir = '/data/SO6/TPOSE/sens_nino34/'
# for a tracer the masks are maskK maskC and maskT
```

```python
# only maskC has more than 1 dimension
Y = 336
X = 1128
maskC = np.fromfile(project_dir + 'nino34_maskC',  dtype='>f4')
maskC = maskC.reshape(Y, X)
maskT = np.fromfile(project_dir + 'nino34_maskT',  dtype='>f4')
maskK = np.fromfile(project_dir + 'nino34_maskK',  dtype='>f4')
print(maskC.shape)
print(maskT.shape)
print(maskK.shape)
```

```python
import matplotlib.pyplot as plt 

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(maskC, cmap='viridis')
plt.show()
```

```python
maskW = np.fromfile('model_setup/surface_transect/maskW.bin', dtype='>f4')
maskW = maskW.reshape(Y, X)
print(maskW.shape)

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(maskW, cmap='viridis')
plt.show()
```

```python
print(np.max(maskC))
print(np.max(maskW))
```
